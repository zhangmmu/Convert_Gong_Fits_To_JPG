```python
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename
from PIL import Image
import matplotlib.pyplot as plt 
import cv2


data0 = fits.open('20100610000154Bh.fits.fz')[0].data
print('There is no data in data0: ', data0)

#compressed image data is in second element of the array
data1 = fits.open('20100610000154Bh.fits.fz')[1].data 
print(data1)
```

```python
#Get image information, the dimension is 2048 * 2048
image_file = get_pkg_data_filename('20100610000154Bh.fits.fz')
fits.info(image_file)
```

```python
plt.grid(None) 
plt.axis('off')

plt.imshow(data1, cmap='gray') 

plt.title('FITS Image') 
#plt.figure(figsize=(5.12, 5.12), dpi=100)

im_resized = cv2.resize(data1, (512, 512), interpolation=cv2.INTER_LINEAR) 
plt.imshow(im_resized, cmap='gray') 
plt.savefig("20100610000154Bh.jpg") 
print(im_resized.shape[0], im_resized.shape[1])
plt.show()  
```
