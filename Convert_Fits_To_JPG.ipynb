```python
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename
from PIL import Image
import matplotlib.pyplot as plt 
import numpy as np
import cv2  

request_url = "https://gong2.nso.edu/HA/haf/201006/20100601/20100601000054Bh.fits.fz"

#compressed image data is in second element of the array, element 0 [0] has no data
original_image_data = fits.open(request_url)[1].data 
```

```python
#Get sample image information, the dimension is 2048 * 2048
image_file = get_pkg_data_filename('20100610000154Bh.fits.fz') 
fits.info(image_file)
```

```python
#resize file to (512,512)
im_resized = cv2.resize(original_image_data, (512, 512), interpolation=cv2.INTER_LINEAR)  
print(im_resized.shape[0], im_resized.shape[1])

plt.grid(None) 
plt.axis('off')  
plt.imshow(im_resized, cmap='gray') 

plt.show()  
```

```python
#save file as jpg file
plt.imsave("20100610000154Bh.jpg", im_resized, cmap='gray') 
```
